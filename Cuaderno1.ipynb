```python
df_c=df_cepre[(df_cepre['ANIO_POSTULA']>2020) & (df_cepre['ANIO_POSTULA']<2024)]
df_c
```

Cuantos postularon por colegio, cuantos ingresaron y cuantos no ingresaron:

```python
def ingreso(x):
  """
  Función para contar la cantidad de "SI" y "NO" en una serie.
  """
  counts = x.value_counts().to_dict()
  return pd.Series({'SI': counts.get('SI', 0), 'NO': counts.get('NO', 0)})

df_a = df_a.groupby(['COLEGIO','COLEGIO_DEPA','COLEGIO_DIST','ESPECIALIDAD'])['INGRESO'].apply(ingreso).unstack() #campo especialidad
df_a = df_a.reset_index()
df_a
```

```python
!pip install fuzzywuzzy
#Import the fuzz module from the fuzzywuzzy package
from fuzzywuzzy import fuzz

def group_similar_colegios(df, threshold=90):
  """
  Groups similar colegio names in a DataFrame using fuzzy string matching.

  Args:
    df: DataFrame with a 'COLEGIO' column.
    threshold: The similarity score threshold for grouping.

  Returns:
    A dictionary where keys are representative colegio names and values are lists of similar colegio names.
  """
  colegios = df.index.get_level_values('COLEGIO').unique().tolist()
  groups = {}
  for colegio in colegios:
    found_group = False
    for group_rep, group_members in groups.items():
      if fuzz.ratio(colegio, group_rep) >= threshold:
        groups[group_rep].append(colegio)
        found_group = True
        break
    if not found_group:
      groups[colegio] = [colegio]
  return groups

# Example usage:
similar_colegios = group_similar_colegios(df_c)

# Print the grouped colegios
for rep_colegio, similar_colegios_list in similar_colegios.items():
  print(f"Representative Colegio: {rep_colegio}")
  print(f"Similar Colegios: {similar_colegios_list}")
  print("-" * 20)
```
